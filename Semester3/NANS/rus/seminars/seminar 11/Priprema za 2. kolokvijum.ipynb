# Подготовка ко второму компьютерному коллоквиуму

**Задание 1.** <br>
Загрузите временной ряд количества пассажиров из файла: [data/lowcost-airline-passengers.csv](data/lowcost-airline-passengers.csv). Эта недорогая компания добивается большого пассажиропотока, но не особо заботится о данных. Сначала замените пропущенные значения, используя модель AR. Затем разделите набор данных на обучение и тестирование в соотношении 80:20 и найдите лучшую модель ARIMA, которая прогнозирует количество пассажиров. Выведите показатель MSE, которого достигает модель. Наконец, нанесите на график измененные временные ряды и прогнозы модели ARIMA (по тестовому набору).

Предположим, что достоверность 95% для всех статистических тестов, если не указано иное.

**Задание 2.** <br>
Загрузите данные музыки и фильма из файла: [data/responses.csv](data/responses.csv) (данные музыки и фильма — это первый 31 столбец).
Разделите набор данных на обучение и тестирование в соотношении 80:20.
Затем примените PCA к данным, сохраняя дисперсию в диапазоне от 49% до 53%.
Используя преобразованные данные, создайте регрессию, которая прогнозирует атрибут «Фильмы» (сколько людей любят смотреть фильмы).
Наконец, выведите скорректированную меру $r^2$, которую достигает модель.
Интерпретируйте два наиболее влиятельных атрибута регрессионной модели.

Предположим, что достоверность 95% для всех статистических тестов, если не указано иное.

*Примечание: используйте «random state = 42» (чтобы результаты были воспроизводимы).*
